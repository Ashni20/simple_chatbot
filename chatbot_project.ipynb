```python
# Simple Rule-Based Chatbot

def chatbot():
    print("Bot: Hello! Type 'bye' to exit.")

    while True:
        user = input("You: ").lower()

        if user == "bye":
            print("Bot: Goodbye!")
            break

        elif "hello" in user or "hi" in user:
            print("Bot: Hi there!")

        elif "how are you" in user:
            print("Bot: I'm just a bot, but I'm doing great!")

        elif "your name" in user:
            print("Bot: I am your AI chatbot.")

        elif "ai" in user:
            print("Bot: AI stands for Artificial Intelligence.")

        else:
            print("Bot: Sorry, I don't understand.")

chatbot()
```

```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# Load Dataset
df = pd.read_csv("chatbot_data.csv")

# Ensure your CSV has these columns: 'question' and 'answer'
questions = df['question'].tolist()
answers = df['answer'].tolist()

# Initialize TF-IDF Vectorizer
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(questions)

# Chatbot Function
def chatbot_response(user_input):
    # Transform user input
    user_vec = vectorizer.transform([user_input])

    # Compute cosine similarity
    similarity = cosine_similarity(user_vec, X)

    # Find index of the most similar question
    idx = np.argmax(similarity)

    # Check if similarity is reasonably high
    if similarity[0][idx] < 0.1:
        return "Sorry, I didn't understand that. Can you ask differently?"
    else:
        return answers[idx]

# Run Chatbot
print("🤖 Chatbot is ready! Type 'exit' to quit.\n")

while True:
    user_input = input("You: ").lower().strip()
    if user_input == "exit" or user_input == "bye":
        print("Chatbot: Goodbye! 👋")
        break
    response = chatbot_response(user_input)
    print("Chatbot:", response)
```
